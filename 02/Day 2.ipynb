```python
string = 'abadfjkdl;fjaskl'

def create_word_dict(string):
    result = {}
    for char in string:
        if char in result:
            result[char] += 1
        else:
            result[char] = 1
    return result

def create_sum_dict(word_dict):
    result = {}
    for k in word_dict:
        if word_dict[k] == 2:
            result[2] = 1
        elif word_dict[k] == 3:
            result[3] = 1
        else:
            continue
    return result        
```

```python
import pandas as pd
df = pd.read_csv('input.txt', header=None)
word_string = df[0].tolist()
```

```python
final = {2:0, 3:0}

for word in word_string:
    word_dict = create_word_dict(word)
    sum_dict = create_sum_dict(word_dict)
    for k in sum_dict:
        final[k] += 1

checksum = final[2] * final[3]
checksum
```

```python
from difflib import SequenceMatcher
# SequenceMatcher(None, 'abcdefgh', 'abcdefhh').ratio()

new_df = pd.DataFrame(columns=word_string)
new_df['Words1'] = word_string
melted_df = pd.melt(new_df, id_vars=['Words1'], 
                    value_name='Ratio', 
                    var_name='Words2')

melted_df.head()
```

```python
for row in melted_df.iterrows():
#     row[2] = SequenceMatcher(None, row[0], row[1]).ratio()
    row[1]
```

```python
def calc_ratio(row):
    return SequenceMatcher(None, row['Words1'], row['Words2']).ratio()

melted_df['Ratio'] = melted_df[['Words1', 'Words2']].apply(calc_ratio, axis=1)
```

```python
new_words_df = melted_df[melted_df['Ratio'] != 1.0]
new_words_df[new_words_df['Ratio'] > 0.96153]
```
