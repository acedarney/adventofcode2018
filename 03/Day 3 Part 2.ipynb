```python
class Rectangle:
    def __init__(self, upper_left, lower_right):
        self.upper_left = upper_left
        self.lower_right = lower_right
        
    def __str__(self):
        return str((self.upper_left, self.lower_right))
    
    def get_corners(self):
        return [self.upper_left, 
                (self.upper_left[0], self.lower_right[1]), 
                (self.lower_right[0], self.upper_left[1]), 
                self.lower_right]
    
    def is_inside(self, other_rect):
        for point in other_rect.get_corners():
            if self.upper_left[0] <= point[0] <= self.lower_right[0] and self.upper_left[1] >= point[1] >= self.lower_right[1]:
                return 1
        return 0
        
```

```python
# test_string = '#1 @ 338,764: 20x24'

def parse_string(input_string):
    decomp1 = input_string.split('@')[1]
    coordinates = decomp1.split(':')[0].strip()
    size = decomp1.split(':')[1].strip()
    loc = (int(coordinates.split(',')[0]), int(coordinates.split(',')[1]))
    size = (int(size.split('x')[0]), int(size.split('x')[1]))
    return loc, size

def get_corners(loc, size):
    return (loc, (loc[0]+size[0], loc[1]-size[1]))
```

```python
rect_dict = {}
ix = 1

with open('input.txt') as f:
    for line in f:
        loc, size = parse_string(line)
        ul, lr = get_corners(loc, size)
        rect_dict[ix] = Rectangle(ul, lr)
        ix += 1
        
# print(rect_dict[2])
```
