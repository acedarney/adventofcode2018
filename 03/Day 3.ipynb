```python
import pandas as pd
```

```python
fabric = pd.DataFrame(index=range(2000), columns=range(2000))
fabric.fillna(int(0), inplace=True)
fabric.head()
```

```python
#1 @ 338,764: 20x24
# ID, x, y of top left corner, height x width of patch

loc = (1, 3)
size = (3, 3)
def add_claim(fabric, loc, size):
    fabric.iloc[loc[1]:loc[1]+size[1], loc[0]:loc[0]+size[0]] += 1
    return fabric
```

```python
# test_string = '#1 @ 338,764: 20x24'

def parse_string(input_string):
    decomp1 = input_string.split('@')[1]
    coordinates = decomp1.split(':')[0].strip()
    size = decomp1.split(':')[1].strip()
    loc = (int(coordinates.split(',')[0]), int(coordinates.split(',')[1]))
    size = (int(size.split('x')[0]), int(size.split('x')[1]))
    return loc, size
```

```python
with open('input.txt') as f:
    for line in f:
        loc, size = parse_string(line)
        add_claim(fabric, loc, size)
        
fabric
```
